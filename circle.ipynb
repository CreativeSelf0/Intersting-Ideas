```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
x = np.linspace(-1,1,100)
x_2 = x**2
d_s = np.sqrt(1 - x_2)
print(len(d_s))
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.scatter(x,d_s)
ax.scatter(x,-d_s)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.grid()
plt.show()
```
